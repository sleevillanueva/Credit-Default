```python
import pandas as pd
import numpy as np
```

```python
cc = pd.read_csv("../assets/CreditDefault.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cc_temp = cc[cc.BILL_AMT2.isnull()!=True]
```

```python
sns.distplot(cc_temp.BILL_AMT2.values)
```

```python
#use median to fill our data based on our positively skewed data
med = np.median(cc_temp.BILL_AMT2.values)
cc.BILL_AMT2.fillna(value=med,inplace=True)
```

```python
#dropping 5,6,and 0 (total of 1% of our data -- drop them)
cc = cc[cc.EDUCATION.isin([1.0,2.0,3.0,4.0])]
```

```python
#same for marriage
cc = cc[cc.MARRIAGE.isin([1,2,3])]
```

```python
#replace -2 to -1 and 0 to -1 for Pay columns
def rep(x):
    x.replace({0:-1, -2:-1}, inplace=True)
```

```python
for i in cc.iloc[:,6:12]:
    rep(cc[i])
```

```python
cc_temp = cc_temp[cc_temp.EDUCATION.isnull() != True]
ed_med = np.median(cc_temp.EDUCATION.values)
```

```python
cc.EDUCATION.fillna(value=ed_med, inplace=True)
```

```python
#changing gender to 0 and 1
cc['SEX'] = cc.SEX.map({2:0, 1:1})
#map marriage and edu as their values
cc["MARRIAGE"] = cc.MARRIAGE.map({1:"m", 2:"s", 3:"o"})
cc["EDUCATION"] = cc.EDUCATION.map({1:"ms", 2:"uni", 3:"hs", 4:"others"})
```

```python
sns.heatmap(cc.corr())
```

```python
dum1 = pd.get_dummies(cc["EDUCATION"])
dum1.drop("uni", axis=1, inplace=True)
```

```python
cc_new = pd.concat([dum1,cc], axis=1)
```

```python
dum2 = pd.get_dummies(cc["MARRIAGE"])
dum2.drop("s", axis=1, inplace=True)
```

```python
cc_new = pd.concat([dum2,cc_new], axis=1)
```

```python
cc_new.drop(["EDUCATION", "MARRIAGE"], axis=1, inplace=True)
```

```python
cc_new.drop("ID", axis=1,inplace=True)
```

```python
pay0 = pd.get_dummies(cc["PAY_0"],prefix="PAY0_")
pay0.drop("PAY0__-1", axis=1, inplace=True)

pay2 = pd.get_dummies(cc["PAY_2"],prefix="PAY2_")
pay2.drop("PAY2__-1", axis=1, inplace=True)

pay3 = pd.get_dummies(cc["PAY_3"],prefix="PAY3_")
pay3.drop("PAY3__-1", axis=1, inplace=True)

pay4 = pd.get_dummies(cc["PAY_4"],prefix="PAY4_")
pay4.drop("PAY4__-1", axis=1, inplace=True)

pay5 = pd.get_dummies(cc["PAY_5"],prefix="PAY5_")
pay5.drop("PAY5__-1", axis=1, inplace=True)

pay6 = pd.get_dummies(cc["PAY_6"],prefix="PAY6_")
pay6.drop("PAY6__-1", axis=1, inplace=True)
```

```python
cc_new = pd.concat([pay0,pay2,pay3,pay4,pay5,pay6,cc_new], axis=1)
```

```python
cc_new.drop(["PAY_0", "PAY_2","PAY_3","PAY_4","PAY_5","PAY_6"], axis=1, inplace=True)
```

```python
from sklearn import preprocessing
from sklearn.cross_validation import train_test_split
from sklearn.grid_search import GridSearchCV
from sklearn import datasets, linear_model
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.cross_validation import StratifiedKFold
from sklearn.feature_selection import RFECV
```

```python
all_columns = "" + "," .join(cc_new.columns - ["default payment next month"]) + ""
all_columns = all_columns.split(",")

xs = cc_new[all_columns]
ys = cc_new["default payment next month"]

xs = preprocessing.scale(xs)

X_train, X_test, y_train, y_test = train_test_split(xs, ys, random_state=1)

lr = linear_model.LogisticRegression()
lr.fit(X_train, y_train)
predictions = lr.predict(X_test)

print accuracy_score(y_test, predictions)
```

```python
rfecv = RFECV(estimator=lr, scoring='accuracy')

# Fit the features to the response variable
X_new = rfecv.fit_transform(xs, ys)

X_train, X_test, y_train, y_test = train_test_split(X_new, ys, random_state=1)

rfecv.fit(X_train, y_train)
predictions = rfecv.predict(X_test)

print rfecv.support_

print accuracy_score(y_test, predictions)
```

```python
y_pred = rfecv.predict(X_new)
print accuracy_score(y_pred,ys)
```

```python
Cs = np.logspace(0.0, 5.0, 20)

parameters = {"penalty": ['l1','l2'],
                     "C":Cs,}

grid = GridSearchCV(lr, parameters,scoring="accuracy")
#grid = GridSearchCV(pipe, param_grid, cv=skf, scoring='roc_auc')
gridmod = grid.fit(X_train, y_train)
y_pred = grid.predict(X_test)
print grid.best_params_
```
